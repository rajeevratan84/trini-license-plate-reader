```python
from os import listdir
from os.path import isfile, join
import os
import cv2

def makeDir(path):
    if os.path.exists(path):
        pass
    else:
        os.makedirs(path)

output_dir = './RENAMED/'       
makeDir(output_dir)
mypath = "./FINAL_PLATES/"
files = [ f.path for f in os.scandir(mypath) if f.is_dir() ]

onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]

labels = []
width = 200
height = 50
dim = (width, height)

for file in onlyfiles:
    image = cv2.imread(mypath+file)
    label = file.split('_')[0]
    label = label.upper()
    try:
        extension = file.split('_')[1].split('.')[1]
        new_file_name = f'{output_dir}{label}.jpg'
        if checkLabel(new_file_name):
            labels.append(new_file_name)
            # resize image
            resized = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)
            cv2.imwrite(new_file_name, resized)
    except:
        pass
```

```python
def checkLabel(label):
    not_allowed = ['(',')',' ','.']
    l = label.split('/')[2][:-4]
    flag_good = True
    for c in l:
        if c in not_allowed:
            print(f"{label} is Bad label")
            flag_good = False
    return flag_good
```
